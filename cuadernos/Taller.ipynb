```python
import pandas as pd
import requests

# URL del conjunto de datos de matr√≠culas
url_matriculas = "https://www.datos.gov.co/resource/nudc-7mev.json?$limit=50000"

# Descarga y conversi√≥n a DataFrame
response = requests.get(url_matriculas)
response.raise_for_status()

df_raw = pd.DataFrame(response.json())
print(f"‚úÖ Datos extra√≠dos: {len(df_raw)} registros")
display(df_raw.head())
```

```python
# URL del conjunto de municipios con c√≥digo DANE
url_municipios = "https://www.datos.gov.co/resource/gdxc-w37w.json?$limit=5000"

# Descarga y conversi√≥n a DataFrame
response_dane = requests.get(url_municipios)
response_dane.raise_for_status()

df_municipios = pd.DataFrame(response_dane.json())
print(f"‚úÖ Municipios cargados: {len(df_municipios)} registros")
display(df_municipios.head())
```

```python
# Copias limpias
df_raw_limpio = df_raw.copy()
df_municipios_limpio = df_municipios.copy()

# Funci√≥n de normalizaci√≥n
def normalizar_texto(col):
    return (
        col.str.strip()
           .str.lower()
           .str.normalize("NFKD")
           .str.encode("ascii", errors="ignore")
           .str.decode("utf-8")
    )

# Normalizar columnas en df_raw
df_raw_limpio["municipio"] = normalizar_texto(df_raw_limpio["municipio"])
df_raw_limpio["departamento"] = normalizar_texto(df_raw_limpio["departamento"])

# Normalizar columnas en df_municipios
df_municipios_limpio["nom_mpio"] = normalizar_texto(df_municipios_limpio["nom_mpio"])
df_municipios_limpio["dpto"] = normalizar_texto(df_municipios_limpio["dpto"])
```

```python
# Uni√≥n de los datos educativos con los c√≥digos oficiales del DANE
df_integrado = pd.merge(
    df_raw_limpio,
    df_municipios_limpio,
    left_on=["departamento", "municipio"],
    right_on=["dpto", "nom_mpio"],
    how="left"
)

# Verificamos si hubo empates exitosos
print(f"üîó Filas unidas: {df_integrado['cod_mpio'].notnull().sum()} de {len(df_integrado)}")
display(df_integrado[["departamento", "municipio", "cod_dpto", "cod_mpio"]].head())
```

```python
import sqlite3

# Ruta a la base de datos (ajusta si es necesario)
ruta_db = r"C:\DIPLOMADO\Datos\educacion_dw.db"

# Crear conexi√≥n
conn = sqlite3.connect(ruta_db)
cursor = conn.cursor()

print("‚úÖ Conexi√≥n establecida.")
```

```python
# Borrar si ya existe
conn.execute("DROP TABLE IF EXISTS Dim_Tiempo;")

# Crear tabla de dimensi√≥n de tiempo
query = """
CREATE TABLE Dim_Tiempo AS
SELECT 
    ROW_NUMBER() OVER (ORDER BY ano) AS id_tiempo,
    ano
FROM (
    SELECT DISTINCT ano
    FROM df_limpio
);
"""

conn.execute(query)
conn.commit()
print("‚úÖ Dim_Tiempo creada correctamente con SQL")

# Mostrar contenido
df_tiempo = pd.read_sql_query("SELECT * FROM Dim_Tiempo", conn)
display(df_tiempo)
```

```python
# Eliminar si existe
conn.execute("DROP TABLE IF EXISTS Dim_Departamento;")

# Crear dimensi√≥n departamento
query_dpto = """
CREATE TABLE Dim_Departamento AS
SELECT 
    ROW_NUMBER() OVER () AS id_departamento,
    cod_dpto,
    nom_dpto AS nombre_departamento
FROM (
    SELECT DISTINCT cod_dpto, nom_dpto
    FROM df_limpio
);
"""

conn.execute(query_dpto)
conn.commit()
print("‚úÖ Dim_Departamento creada correctamente.")
```

```python
# Eliminar si existe
conn.execute("DROP TABLE IF EXISTS Dim_Municipio;")

# Crear dimensi√≥n municipio con FK a departamento
query_mpio = """
CREATE TABLE Dim_Municipio AS
SELECT 
    ROW_NUMBER() OVER () AS id_municipio,
    cod_mpio,
    nom_mpio AS nombre_municipio,
    cod_dpto  -- FK l√≥gica
FROM (
    SELECT DISTINCT cod_mpio, nom_mpio, cod_dpto
    FROM df_limpio
);
"""

conn.execute(query_mpio)
conn.commit()
print("‚úÖ Dim_Municipio creada correctamente.")
```

```python
# üßΩ Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS Hechos_Educacion;")

# üìä Crear tabla de hechos
query = """
CREATE TABLE Hechos_Educacion AS
SELECT 
    t.id_tiempo,
    m.id_municipio,
    l.poblaci_n_5_16,
    l.tasa_matriculaci_n_5_16,
    l.cobertura_neta,
    l.aprobaci_n,
    l.reprobaci_n,
    l.repitencia,
    l.deserci_n
FROM df_limpio l
JOIN Dim_Tiempo t ON l.ano = t.ano
JOIN Dim_Municipio m 
    ON l.cod_mpio = m.cod_mpio 
   AND l.cod_dpto = m.cod_dpto;
"""

conn.execute(query)
conn.commit()
print("‚úÖ Tabla Hechos_Educacion actualizada con id_municipio correctamente.")
```

```python
query1 = """
SELECT 
    m.nombre_municipio AS nombre_municipio,
    m.cod_mpio AS codigo_municipio,
    h.poblaci_n_5_16,
    h.tasa_matriculaci_n_5_16,
    ROUND(h.tasa_matriculaci_n_5_16 * 100, 2) AS porcentaje_escolaridad
FROM Hechos_Educacion h
JOIN Dim_Municipio m ON h.id_municipio = m.id_municipio
ORDER BY porcentaje_escolaridad DESC;
"""

df_porcentaje_escolaridad = pd.read_sql_query(query1, conn)
df_porcentaje_escolaridad.head()
```

```python
query_rendimiento = """
SELECT 
    m.nombre_municipio AS municipio,
    ROUND(AVG(h.aprobaci_n), 2) AS aprobacion_promedio,
    ROUND(AVG(h.reprobaci_n), 2) AS reprobacion_promedio
FROM Hechos_Educacion h
JOIN Dim_Municipio m ON h.id_municipio = m.id_municipio
GROUP BY m.nombre_municipio
ORDER BY aprobacion_promedio DESC, reprobacion_promedio ASC;
"""

df_rendimiento = pd.read_sql_query(query_rendimiento, conn)
display(df_rendimiento)
```

```python
query_cobertura = """
SELECT 
    d.nombre_departamento,
    ROUND(AVG(h.cobertura_neta), 2) AS cobertura_promedio
FROM Hechos_Educacion h
JOIN Dim_Municipio m ON h.id_municipio = m.id_municipio
JOIN Dim_Departamento d ON m.cod_dpto = d.cod_dpto
GROUP BY d.nombre_departamento
ORDER BY cobertura_promedio DESC;
"""

df_cobertura = pd.read_sql_query(query_cobertura, conn)
display(df_cobertura)
```
