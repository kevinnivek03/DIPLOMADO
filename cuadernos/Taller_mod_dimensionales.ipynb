La base DIVIPOLA (Divisi√≥n Pol√≠tico-Administrativa de Colombia) es la fuente oficial del DANE encargada de codificar y organizar los distintos niveles territoriales del pa√≠s. Proporciona c√≥digos y nombres estandarizados para departamentos, municipios, corregimientos, centros poblados y cabeceras municipales, facilitando la homogeneizaci√≥n geogr√°fica en conjuntos de datos gubernamentales. Cada municipio cuenta con un c√≥digo √∫nico de seis cifras: los dos primeros d√≠gitos corresponden al departamento y los cuatro siguientes al municipio.

Esta base ser√° empleada para construir una de las dimensiones de nuestro modelo de datos.

```python
import unidecode

# Copia de trabajo
df = df_raw.copy()

# 1. Eliminar columnas completamente vac√≠as
df = df.dropna(axis=1, how='all')

# 2. Renombrar columnas: quitar tildes, espacios, may√∫sculas
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace(r'[^a-z0-9_]', '', regex=True)
    .map(unidecode.unidecode)  # quitar tildes
)

# 3. Eliminar filas del departamento 'NACIONAL' o similares
df = df[~df['departamento'].str.upper().str.strip().isin(['NACIONAL'])]

# 4. Normalizar texto de municipio y departamento
df['municipio'] = df['municipio'].astype(str).str.strip().str.upper().map(unidecode.unidecode)
df['departamento'] = df['departamento'].astype(str).str.strip().str.upper().map(unidecode.unidecode)

# 5. Convertir columnas num√©ricas (las que se puedan)
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')

# 6. Asegurar que la columna de a√±o est√© como entero
if 'ano' in df.columns:
    df['ano'] = pd.to_numeric(df['ano'], errors='coerce').astype('Int64')

# 7. Resetear √≠ndice
df.reset_index(drop=True, inplace=True)

# Guardar en session_state o variable limpia
df_clean = df.copy()
print("‚úÖ Transformaci√≥n completa aplicada correctamente.")
```

```python
# Copia de trabajo
df_geo = df_divipola.copy()

# 1. Normalizar columnas
df_geo.columns = (
    df_geo.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace(r'[^a-z0-9_]', '', regex=True)
)

# 2. Limpiar y estandarizar nombres
df_geo['dpto'] = (
    df_geo['dpto']
    .astype(str)
    .str.strip()
    .str.upper()
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
)

df_geo['nom_mpio'] = (
    df_geo['nom_mpio']
    .astype(str)
    .str.strip()
    .str.upper()
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
)

# 3. Rellenar ceros en c√≥digos
df_geo['cod_dpto'] = df_geo['cod_dpto'].astype(str).str.zfill(2)
df_geo['cod_mpio'] = df_geo['cod_mpio'].astype(str).str.zfill(3)

# 4. Crear ID geogr√°fico √∫nico (6 d√≠gitos)
df_geo['id_geo'] = df_geo['cod_dpto'] + df_geo['cod_mpio']

# 5. Eliminar duplicados
df_geo = df_geo.drop_duplicates(subset='id_geo')

# 6. Resetear √≠ndice
df_geo.reset_index(drop=True, inplace=True)

# 7. Renombrar columnas clave para modelo dimensional
df_geo = df_geo.rename(columns={
    'dpto': 'departamento',
    'nom_mpio': 'municipio'
})

print("‚úÖ Base DIVIPOLA transformada exitosamente y lista para usar como dimensi√≥n geogr√°fica.")
```

## Dimensiones para el modelo estrella

```python
# Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS dim_tiempo;")

# Crear la dimensi√≥n de tiempo
query_dim_tiempo = """
CREATE TABLE dim_tiempo AS
SELECT 
    ROW_NUMBER() OVER () AS id_tiempo,
    a_o AS anio
FROM educacion_raw
WHERE a_o IS NOT NULL
GROUP BY a_o
ORDER BY a_o;
"""

conn.execute(query_dim_tiempo)
```

```python
# Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS dim_municipio;")

# Crear la dimensi√≥n con los nombres exactos de columnas
query_dim_municipio = """
CREATE TABLE dim_municipio AS
SELECT 
    ROW_NUMBER() OVER () AS id_municipio,
    d.id_geo,
    d.cod_dpto AS codigo_departamento,
    d.departamento,
    d.cod_mpio AS codigo_municipio,
    d.municipio,
    d.tipo_municipio,
    d.latitud,
    d.longitud
FROM divipola d
GROUP BY 
    d.id_geo,
    d.cod_dpto,
    d.departamento,
    d.cod_mpio,
    d.municipio,
    d.tipo_municipio,
    d.latitud,
    d.longitud;
"""

conn.execute(query_dim_municipio)
```

```python
# Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS dim_departamento;")

# Crear la tabla con el nombre original exacto de la columna
query_dim_departamento = """
CREATE TABLE dim_departamento AS
SELECT 
    ROW_NUMBER() OVER () AS id_departamento,
    c_digo_departamento AS codigo_departamento,
    departamento
FROM educacion_raw
WHERE c_digo_departamento IS NOT NULL
GROUP BY c_digo_departamento, departamento;
"""

conn.execute(query_dim_departamento)
```

## Tabla de hechos

```python
# Eliminar si ya existe
conn.execute("DROP TABLE IF EXISTS fact_educacion;")

# Crear nueva tabla de hechos con m√©tricas clave
query_fact_educacion = """
CREATE TABLE fact_educacion AS
SELECT 
    t.id_tiempo,
    m.id_municipio,
    d.id_departamento,
    e.tasa_matriculaci_n_5_16 AS tasa_matriculacion,
    e.cobertura_neta,
    e.aprobaci_n AS aprobacion,
    e.reprobaci_n AS reprobacion,
    e.repitencia,
    e.deserci_n AS desercion
FROM educacion_raw e
JOIN dim_tiempo t 
    ON e.a_o = t.anio
JOIN dim_municipio m 
    ON CAST(e.c_digo_municipio AS TEXT) = m.codigo_municipio
JOIN dim_departamento d 
    ON CAST(e.c_digo_departamento AS TEXT) = d.codigo_departamento
WHERE 
    e.tasa_matriculaci_n_5_16 IS NOT NULL
    AND e.cobertura_neta IS NOT NULL;
"""

conn.execute(query_fact_educacion)
```

## 1. Respecto a la poblaci√≥n del municipio ¬øQue porcentaje de escolaridad hay?

```python
query_top_municipios = """
SELECT 
    m.municipio,
    ROUND(AVG(h.tasa_matriculacion), 2) AS tasa_promedio
FROM fact_educacion h
JOIN dim_municipio m 
    ON h.id_municipio = m.id_municipio
GROUP BY m.municipio
HAVING tasa_promedio IS NOT NULL
ORDER BY tasa_promedio DESC
LIMIT 10;
"""

df_top_municipios = pd.read_sql(query_top_municipios, conn)
df_top_municipios
```

### üìä Interpretaci√≥n de los resultados

El an√°lisis de la tasa promedio de matriculaci√≥n por municipio revela diferencias significativas en la cobertura educativa a nivel local. Municipios como **Cota**, **Tenjo** y **La Calera** presentan los valores m√°s altos, lo cual sugiere un amplio acceso a la educaci√≥n en comparaci√≥n con otros territorios. Estos resultados podr√≠an estar relacionados con una buena infraestructura escolar, una mayor oferta educativa o condiciones demogr√°ficas particulares.

En particular, el caso de **Cota**, con una tasa promedio de matriculaci√≥n del **209.95%**, indica que el n√∫mero de estudiantes matriculados es m√°s del doble de la poblaci√≥n estimada entre 5 y 16 a√±os. Este tipo de valores, aunque poco comunes, pueden explicarse por:

- La inclusi√≥n de estudiantes fuera del rango de edad oficial (por rezago o adelantamiento),
- La recepci√≥n de estudiantes provenientes de otros municipios (migraci√≥n escolar),
- O posibles diferencias entre la poblaci√≥n proyectada y la atendida realmente.

Tasas superiores al 100% no necesariamente reflejan un error, pero s√≠ invitan a profundizar en el contexto educativo y demogr√°fico de cada territorio.


## 2. ¬øC√≥mo comparar√≠a el rendimiento educativo por municipios?

```python
query_rendimiento_municipal = """
SELECT 
    m.municipio,
    ROUND(AVG(f.aprobacion), 2) AS promedio_aprobacion,
    ROUND(AVG(f.reprobacion), 2) AS promedio_reprobacion,
    ROUND(AVG(f.repitencia), 2) AS promedio_repitencia,
    ROUND(AVG(f.desercion), 2) AS promedio_desercion
FROM fact_educacion f
JOIN dim_municipio m ON f.id_municipio = m.id_municipio
GROUP BY m.municipio
HAVING promedio_aprobacion IS NOT NULL
ORDER BY promedio_aprobacion DESC
LIMIT 10;
"""

df_rendimiento = pd.read_sql(query_rendimiento_municipal, conn)
df_rendimiento
```

### üéì Municipios con mejor rendimiento educativo

Este an√°lisis muestra los 10 municipios con mayor promedio de aprobaci√≥n escolar. Todos superan el 97.5% de aprobaci√≥n, lo que refleja un alto desempe√±o en t√©rminos de permanencia y logro acad√©mico dentro del sistema educativo.

Municipios como **Puerto Arica**, **Tibirit√°** y **La Palma** encabezan el listado, combinando altos niveles de aprobaci√≥n con niveles bajos de reprobaci√≥n y deserci√≥n. Sin embargo, es importante observar que algunos casos, como **Puerto Arica** y **Miriti - Paran√°**, presentan niveles elevados de repitencia, lo que podr√≠a indicar desaf√≠os en la progresi√≥n escolar a pesar de que los estudiantes finalmente aprueban.

Este tipo de an√°lisis permite identificar territorios con buenas pr√°cticas educativas, pero tambi√©n aquellos donde, a pesar de las altas tasas de aprobaci√≥n, pueden existir patrones de repitencia o abandono que deben ser atendidos para garantizar trayectorias educativas continuas y efectivas.


## 3. ¬øQue departamentos son los que mejor cobertura tienen? ¬øPueden hacer c√°lculo con SQL?

```python
query_cobertura_ponderada_fix = """
SELECT 
    UPPER(REPLACE(REPLACE(d.departamento, ',', ''), '.', '')) AS departamento,
    ROUND(SUM(e.cobertura_neta * e.poblaci_n_5_16) / SUM(e.poblaci_n_5_16), 2) AS cobertura_neta_ponderada
FROM educacion_raw e
JOIN dim_departamento d 
    ON CAST(e.c_digo_departamento AS TEXT) = d.codigo_departamento
WHERE 
    e.cobertura_neta IS NOT NULL 
    AND e.poblaci_n_5_16 IS NOT NULL 
    AND e.poblaci_n_5_16 > 0
GROUP BY UPPER(REPLACE(REPLACE(d.departamento, ',', ''), '.', ''))
ORDER BY cobertura_neta_ponderada DESC
LIMIT 20;
"""

df_resultado = pd.read_sql(query_cobertura_ponderada_fix, conn)
df_resultado
```

### üó∫Ô∏è Departamentos con mejor cobertura educativa (ponderada)

El an√°lisis de cobertura neta ponderada permite identificar los departamentos donde una mayor proporci√≥n de la poblaci√≥n entre 5 y 16 a√±os est√° efectivamente matriculada. A diferencia de un promedio simple, esta m√©trica ajusta la cobertura con base en el tama√±o de la poblaci√≥n infantil, ofreciendo una visi√≥n m√°s justa y representativa.

Los resultados muestran que departamentos como **Casanare (96.97%)**, **Sucre (95.38%)** y **Santander (94.02%)** lideran en cobertura, lo que sugiere una adecuada capacidad del sistema educativo para atender a la poblaci√≥n escolar en esos territorios.

En contraste, aunque **Bogot√° D.C.** se encuentra en el listado, su cobertura ponderada es del **88.63%**, lo que indica oportunidades de mejora, especialmente considerando su rol como capital y centro educativo nacional.

Este tipo de an√°lisis es √∫til para priorizar intervenciones y monitorear el desempe√±o regional en t√©rminos de acceso efectivo a la educaci√≥n.
